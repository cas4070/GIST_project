```python
#python xml library import 
import xml.etree.ElementTree as ET

#폴더 내 파일 이름 얻기 위한 library
import os

#판다스
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 폴더 내의 모든 파일 이름 얻어오기.
file_name = []
for root, dirs, files in os.walk("/home/share/aschoi/nas/users/asolchoi/data/COCONUT_Herbal/4_HMDB/hmdb_metabolites/"):
    for f in files:
        file_name.append(f)

# HMDB00001 부터 시작하게 sort한다.
file_name.sort()

# 마지막은 전체 파일이기 때문에 삭제 한다.
remove_index = len(file_name)-1
del file_name[remove_index]


#file별 path
file_path=[]

for i in file_name:
    path = "/home/share/aschoi/nas/users/asolchoi/data/COCONUT_Herbal/4_HMDB/hmdb_metabolites/" + i
    file_path.append(path)
```

```python
# 결과 저장할 dataframe 생성

df = pd.DataFrame(columns=['Accession','Name','PubChem_CID'])
df
```

```python
# xml 파일별로 필요한 정보 파싱
cnt = 0 
for j in file_path:
    #parse xml file
    tree = ET.parse(j)
    #get root node
    xml_root = tree.getroot()
    
    #필요한 데이터 저장
    data = []
    data.append(xml_root[3].text) #accession
    if type(xml_root[5].text) == unicode: #name
        data.append(xml_root[5].text.encode('ascii', 'replace'))
    else:
        data.append(xml_root[5].text) 
    data.append(xml_root[43].text) #PubChem CID
        
    #dataframe 생성
    temp_df =  pd.DataFrame([data], columns=['Accession','Name','PubChem CID'])
    
    #결과가 저장될 dataframe에 추가
    df = df.append(temp_df)
    
    cnt = cnt +1
    if cnt % 10000 == 0 :
        print(cnt)
    
    
```

```python
df.rename(columns={'PubChem CID' : 'PubChem_CID'}, inplace=True)
```

```python
#cid 존재하는 애들
hmdb_cid = df[~df.PubChem_CID.isnull()]
```

```python
hmdb_cid.to_csv("/home/share/aschoi/nas/users/asolchoi/data/COCONUT_Herbal/4_HMDB/HMDB_PubChem_cid.tsv", sep='\t', index=False )
```
