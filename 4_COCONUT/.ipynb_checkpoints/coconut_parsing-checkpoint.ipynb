```python
#접속하기
db_handle = MySQLdb.connect('localhost', 'root','comsysbio', 'COCONUTv5')
cur = db_handle.cursor(MySQLdb.cursors.DictCursor) # query를 통해 데이터를 얻을 때 python의 dict로 얻기 위한 방법
cursor = db_handle.cursor()
```

```python
query1 = "SELECT herbID, name FROM Herb"
cursor.execute(query1); #쿼리 날리는 거
results = cursor.fetchall() # cursor.fetchone()은 tuple, cursor.fetchall()은 list of tuples을 리턴한다.
for row in results:
    print row
```

```python
#tuple type
tup_temp = []
for row in tup_results:
    herb_id = row[0]
    herb_name = row[1]
    query2 = "SELECT compID, name, SMILES, InChI From Compound WHERE compID IN(SELECT compID FROM Herb_Compound WHERE herbID= %ld)" % herb_id
    cursor.execute(query2)
    results = cursor.fetchall()
    tup_temp.append([herb_id, herb_name, results])
```

```python
#dictionary type
dict_temp = []
for row in tup_results:
    herb_id = row[0]
    query2 = "SELECT compID, name, SMILES, InChI From Compound WHERE compID IN(SELECT compID FROM Herb_Compound WHERE herbID= %ld)" % herb_id
    cur.execute(query2)
    results = cur.fetchall()
    dict_temp.append(results)
```

```python
#결과 table 만들기
df = pd.DataFrame(columns=['compID','compName','smiles', 'InChI','herbID','name'])
for i in tup_temp:
    for j in i[2]: #several COCONUT compounds were related to one HerbID
        temp_df=pd.DataFrame([[j[0], j[1], j[2], j[3], i[0], i[1]]], columns=['compID','compName','smiles', 'InChI','herbID','name'])
        df = df.append(temp_df)
```

```python
# save file
df.to_csv("/home/csbl/nas/users/asolchoi/data/COCONUT_Herbal/4_HMDB/COCONUT_herb_compound.csv", mode='w')
```

```python
# make dataframe index 0 to len(df)
df.index = range(0,len(df))
```

```python
# InChI and smiles value has null at the same time.

# get index (InChI and smiles value has null at the same time.)
rm_index = df[((df.InChI == 'null') | (df.InChI == 'NULL')) & ((df.smiles == 'null') | (df.smiles =='NULL'))].index.tolist()

#remove such dataframe
df = df.drop(rm_index)
```

```python
# save file
df.to_csv("/home/csbl/nas/users/asolchoi/data/COCONUT_Herbal/4_HMDB/COCONUT_herb_compound(remove_null).csv", mode='w')
```

```python
## compound 중복 없애기(compID 기준)
df_duplicate_dropped = df.drop_duplicates(subset='compID')
```

```python
df_duplicate_dropped
```

```python
# get herbID and name about one compID
unique_list = df.compID.unique().tolist()
compID_herbID = []
compID_herbName = []
for i in unique_list:
    tdf = df[(df.compID == i)]
    herb_id_of_compID = list(tdf['herbID'])  ##series type
    herb_name_of_compID = list(tdf['name']) ## series type
    compID_herbID.append(herb_id_of_compID)
    compID_herbName.append(herb_name_of_compID)       
```
