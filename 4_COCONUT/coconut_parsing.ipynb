```python
#접속하기
db_handle = MySQLdb.connect('localhost', 'root','comsysbio', 'COCONUTv5')
cur = db_handle.cursor(MySQLdb.cursors.DictCursor) # query를 통해 데이터를 얻을 때 python의 dict로 얻기 위한 방법
cursor = db_handle.cursor()
```

```python
query1 = "SELECT herbID, name, reference FROM Herb"
#query1 = "SELECT herbID, name FROM Herb WHERE NOT(reference LIKE '')"
cursor.execute(query1); #쿼리 날리는 거
results = cursor.fetchall() # cursor.fetchone()은 tuple, cursor.fetchall()은 list of tuples을 리턴한다.
for row in results:
    print row
```

```python
tup_results = results
```

```python
print len(tup_results)
```

```python
##############Herb phenotype############
#tuple type
herb_temp = []
for row in tup_results:
    herb_id = row[0]
    herb_name = row[1]
    query2 = "SELECT usePart, evidence, reference From Herb_Phenotype WHERE herbID= %ld" % herb_id
    cursor.execute(query2)
    results = cursor.fetchall()
    herb_temp.append([herb_id, herb_name, results])
```

```python
#결과 table 만들기
df = pd.DataFrame(columns=['herbID','herbName','usePart', 'evidence', 'reference'])
for i in herb_temp:
    for j in i[2]:
        temp_df=pd.DataFrame([[i[0], i[1], j[0], j[1], j[2]]],columns=['herbID','herbName','usePart', 'evidence', 'reference'])
        df = df.append(temp_df)
```

```python
# make dataframe index 0 to len(df)
df.index = range(0,len(df))
df.to_csv("/home/csbl/nas/users/asolchoi/data/COCONUT_Herbal/4_HMDB/COCONUT_herb_phenotype3.csv", mode='w')
```

```python
df_duplicate = df.drop_duplicates(subset=['herbName','evidence'])
df_duplicate.to_csv("/home/csbl/nas/users/asolchoi/data/COCONUT_Herbal/4_HMDB/COCONUT_herb_phenotype3(remove duplicate).csv")
```

```python
########################Herb compound #####################
#tuple type
comp_temp = []
for row in tup_results:
    herb_id = row[0]
    herb_name = row[1]
    herb_ref = row[2]
    query2 = "SELECT compID, name, SMILES, InChI, reference From Compound WHERE compID IN(SELECT compID FROM Herb_Compound WHERE herbID= %ld)" % herb_id
    cursor.execute(query2)
    results = cursor.fetchall()
    comp_temp.append([herb_id, herb_name, herb_ref, results])
```

```python
#결과 table 만들기
df2 = pd.DataFrame(columns=['compID','compName','smiles', 'InChI','compRef','herbID','name', 'herbRef'])
for i in comp_temp:
    for j in i[3]: #several COCONUT compounds were related to one HerbID
        temp_df2=pd.DataFrame([[j[0], j[1], j[2], j[3],j[4],i[0], i[1], i[2]]], columns=['compID','compName','smiles', 'InChI','compRef','herbID','name', 'herbRef'])
        df2 = df2.append(temp_df2)
```

```python
# make dataframe index 0 to len(df)
df2.index = range(0,len(df2))
# save file
df2.to_csv("COCONUT_herb_compound+ref.csv", mode='w')
```

```python
# get herbID and name about one compID
unique_list = df3.compID.unique().tolist()
compID_herbID = []
compID_herbName = []
compID_herbRef = []
compID_compRef = []
for i in unique_list:
    tdf = df3[(df3.compID == i)]
    herb_id_of_compID = list(tdf['herbID'])  ##series type
    herb_name_of_compID = list(tdf['name']) ## series type
    herb_ref_of_compID = list(tdf['herbRef'])
    comp_ref_of_compID = list(tdf['compRef'])
    compID_herbID.append(herb_id_of_compID)
    compID_herbName.append(herb_name_of_compID)       
    compID_herbRef.append(herb_ref_of_compID)
    compID_compRef.append(comp_ref_of_compID)
```

```python
########################Herb disease association#####################
#tuple type
disease_temp = []
for row in tup_results:
    herb_id = row[0]
    herb_name = row[1]
    query2 = "SELECT * From Phenotype WHERE phenID IN(SELECT phenID FROM Herb_Phenotype WHERE herbID = %ld)" % herb_id
    cursor.execute(query2)
    results = cursor.fetchall()
    disease_temp.append([herb_id, herb_name, results])
```

```python
#결과 table 만들기
df3 = pd.DataFrame(columns=['herbID','herbName','phenID','name','umlsID', 'umlsSemanticType','reference'])
for i in disease_temp:
    for j in i[2]: #several COCONUT compounds were related to one HerbID
        temp_df3=pd.DataFrame([[i[0], i[1], j[0], j[1], j[2], j[3], j[4]]], columns=['herbID','herbName','phenID','name','umlsID', 'umlsSemanticType','reference'])
        df3 = df3.append(temp_df3)
```

```python
df3.to_csv("/home/csbl/nas/users/asolchoi/data/COCONUT_Herbal/4_HMDB/COCONUT_herb_disease.csv", mode='w')
```

```python
####################### disease feature#######################
disease_feature = []
for row in tup_results:
    herb_id = row[0]
    herb_name = row[1]
    query2 = "SELECT phenID, geneID, directEvidence From Phenotype_Gene WHERE phenID IN(SELECT phenID FROM Herb_Phenotype WHERE herbID = %ld)" % herb_id
    cursor.execute(query2)
    results = cursor.fetchall()
    disease_feature.append([herb_id, herb_name, results])
```

```python
#결과 table 만들기
df4 = pd.DataFrame(columns=['herbID','herbName','phenID','geneID', 'DirectEvidence'])
for i in disease_feature:
    for j in i[2]: #several COCONUT compounds were related to one HerbID
        temp_df4=pd.DataFrame([[i[0], i[1], j[0], j[1], j[2]]], columns=['herbID','herbName','phenID','geneID', 'DirectEvidence'])
        df4 = df4.append(temp_df4)
```

```python
df4.to_csv("/home/csbl/nas/users/asolchoi/data/COCONUT_Herbal/4_HMDB/COCONUT_herb_disease_feature1.csv", mode='w')
```
