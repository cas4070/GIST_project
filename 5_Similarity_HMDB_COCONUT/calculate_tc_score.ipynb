```python
def fingerprint(compID, smiles):
    cid=compID
    smi=smiles
    mols = []
    fps = []
    idx = -1

    #### getting mols
    print "getting mols start : " + time.strftime('%a %H:%M:%S') 
    for x in smi:
        idx = idx + 1
        try:
            mols.append(pybel.readstring("smi", x))
        except IOError:
            cid[idx] = 'error'
    print "getting mols end : " + time.strftime('%a %H:%M:%S')
    
    #### caculating fingerprints
    print"calculating fingerprints start : " + time.strftime('%a %H:%M:%S')
    for y in mols :
        fps.append(y.calcfp(fptype='FP2'))
    print"calculating fingerprints end : " + time.strftime('%a %H:%M:%S')
    
    ### remove error elements in cid
    print cid.count('error')
    for i in range(0, cid.count('error')):
        cid.remove('error')
    
    return cid, fps
            
```

```python
### calculate herb compound's fingerprint using its smiles
herb_smiles = herb['smiles'].tolist()
herb_cid =herb['compID'].tolist()
herb_fps = []
herb_cid, herb_fps = fingerprint(herb_cid, herb_smiles)
```

```python
### calculate hmdb compound's fingerprint using its smiles
hmdb_smiles = hmdb['ChemSpider_Smiles'].tolist()
hmdb_cid = hmdb['Accession'].tolist()
hmdb_fps = []
hmdb_cid, hmdb_fps = fingerprint(hmdb_cid, hmdb_smiles)
```

```python
#calculate similarity (tanimotto coefficient)

def tc_score(herb_fps, hmdb_fps):
    score = []
    print"start : " + time.strftime('%a %H:%M:%S')
    for i in herb_fps:
        for j in hmdb_fps:
            score.append(i|j)
    print"end : " + time.strftime('%a %H:%M:%S')
    
    return score
```

```python
tc_vector = []
tc_vector = tc_score(herb_fps, hmdb_fps)
```

```python
# 1D list to 2D lists
def to_matrix(l, n) :
    return [l[i:i + n] for i in xrange(0, len(l), n)]
```

```python
tc_matrix = to_matrix(tc_vector, len(hmdb_cid))
```
