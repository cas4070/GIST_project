```python
# rs가 이중 리스트(안에 있는 것은 tuple)구조로 되어 있어서 벗겨낸다.
temp = list()
for i in rs:
    temp.append(i[0])
```

```python
search_term = list()
cas_kegg = list()
for i in temp:
    s_term = i['searchTerm'].encode('ascii', 'replace')
    res = i['result']
    #str_res = ', '.join(res)
    
    search_term.append(s_term)
    if res == []:
        cas_kegg.append('null')
        continue
    else:
        #cas_kegg.append(str_res.encode('ascii', 'replace'))
        #결과가 여러개 나오기는 하지만, 첫번째에 매치되는 것이 제일 score가 높기 때문에 첫번째 애만 골랐음.
        cas_kegg.append(res[0].encode('ascii', 'replace'))
```

```python
# update PubChem Id from casID
for i, j in zip(search_term, cas_kegg):
    temp_df3 = df2[df2.casID == i]
    idx  = temp_df3.index.values.tolist()
    for k in idx:
        df2.loc[k, 'pubchemID'] = j
```
