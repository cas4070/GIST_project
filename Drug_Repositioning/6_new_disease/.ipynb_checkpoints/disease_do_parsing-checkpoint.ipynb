```python
import pandas as pd
import numpy as np
```

```python
f = open("/home/share/aschoi/nas/users/asolchoi/data/Drug_Repositioning/6_new_disease/HumanDO.obo", 'r')
```

```python
lines = f.readlines()
```

```python
data2 = list()
for line in lines:
    if line.startswith('id: DOID:'):
        data2.append(line)
        continue
    if 'xref: OMIM:' in line:
        data2.append(line)
```

```python
data2
```

```python
def indices(lst, element):
    result = []
    offset = -1
    while True:
        try:
            offset = lst.index(element, offset+1)
        except ValueError:
            return result
        result.append(offset)
```

```python
do_idx = list()
for idx in range(len(data2)):
    if data2[idx].startswith('id: DO'):
        do_idx.append(idx)
```

```python
mim_idx = list()
for idx in range(len(data2)):
    if data2[idx].startswith('xref'):
        mim_idx.append(idx)
```

```python
raw = pd.DataFrame()
i = 0
j = 0
while i < len(do_idx):
    while j < len(mim_idx):
        if (do_idx[i] < mim_idx[j]) & (mim_idx[j] < do_idx[i+1]):
            temp_df = pd.DataFrame(data=[[data2[do_idx[i]], data2[mim_idx[j]]]], columns=['do_id', 'mim_id'])
            raw = raw.append(temp_df)
            j = j+1
        else:
            i = i+1
            break 
```
