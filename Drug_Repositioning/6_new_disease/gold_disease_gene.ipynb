```python
import pandas as pd
import numpy as np
```

```python
f = open("/home/share/aschoi/nas/users/asolchoi/data/Drug_Repositioning/6_new_disease/gold_omim_id.txt", 'r')
gold_di = list()
while True:
    line = f.readline()
    if not line:break
    gold_di.append(int(line.lstrip("OMIM:").rstrip("\n")))
f.close()
```

```python
mim_uml = pd.read_table("/DAS_Storage1/aschoi/data/Drug_Repositioning/6_new_disease/1_omim_umls_mapping_association.tsv")
```

```python
le = mim_uml[mim_uml.OMIM_ID.isin(gold_di)]
```

```python
le.head(1)
```

```python
dis_ge = pd.read_table("/DAS_Storage1/aschoi/data/Drug_Repositioning/6_new_disease/curated_gene_disease_associations.tsv", skiprows=21)
```

```python
dis_ge.diseaseId = dis_ge.diseaseId.str.lstrip("umls:")
```

```python
merged = pd.merge(left= le, right = dis_ge, left_on="UMLS_concept_ID", right_on = 'diseaseId', how='inner')
```

```python
gold_di_feature = merged[["OMIM_ID", "OMIM_disease_name", "UMLS_concept_ID", "UMLS_concept_name", "geneId", "score", "geneName"]]
```
